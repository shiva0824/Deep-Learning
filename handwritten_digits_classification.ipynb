```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train, y_train) , (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
print(len(X_train), len(X_test))
```

```python
y_train[0]
```

```python
#highest pixel value 255
#scaling values between 0 and 1
X_train = X_train / 255
X_test = X_test / 255
```

```python
X_train[0]
```

```python
#flattening our 2d array to 1d array to give it as input to our neural network
X_train_flattened = X_train.reshape(len(X_train), 28*28) #28*28 is size of 2d array
X_test_flattened = X_test.reshape(len(X_test), 28*28)
```

```python
#see the diff....2d array(28*28) is converted to 1d array(which consists 784 values)
print(X_train.shape)
print(X_train_flattened.shape)
```

digits_nn.jpg

```python
#above image is the simple neural netwrok with no hidden layers
```

```python
#model creation

model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])

#model optimizer
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train_flattened, y_train, epochs=5)
```

```python
model.evaluate(X_test_flattened, y_test)
```

```python
plt.matshow(X_test[0])
```

```python
y_predicted = model.predict(X_test_flattened)
y_predicted[0]
```

```python
#np.argmax finds a maximum element from an array and returns index of it
np.argmax(y_predicted[0])
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
#using hidden layer..with activation function 'relu'
#reason: using hidden layer...will better our model performance.

model = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train_flattened, y_train, epochs=5)
```

```python
model.evaluate(X_test_flattened,y_test)
```

```python
y_predicted = model.predict(X_test_flattened)
y_predicted_labels = [np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
#keras has special method to flaten the input dataset...instead of doing it spearately using .reshape method

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=10)
```

```python
model.evaluate(X_test,y_test)
```
