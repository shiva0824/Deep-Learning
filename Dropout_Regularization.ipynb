```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from keras import layers
from sklearn.metrics import confusion_matrix , classification_report
```

```python
'''This is a dataset that describes sonar chirp returns bouncing off different services. 
The 60 input variables are the strength of the returns at different angles. 
It is a binary classification problem that requires a model to differentiate rocks from metal cylinders.
'''
```

```python
df = pd.read_csv('sonar_dataset.csv')
```

```python
df.shape
```

```python
#df.isnull().sum()
```

```python
df.head()
```

```python
df['R'].value_counts()
```

```python
X = df.drop('R', axis = 1)
y = df['R']
```

```python
#X
```

```python
#Y
```

```python
y = pd.get_dummies(y, drop_first=True)
y.sample(5)  #R-->1, M--->0
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
#model
model = keras.Sequential([
    layers.Dense(60, input_dim=60, activation='relu'),
    layers.Dense(30, activation='relu'),
    layers.Dense(15, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model.compile(loss = 'binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
print(classification_report(y_test, y_pred))
```

```python
#Model with Dropout Layer
modeld = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
])

modeld.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

modeld.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
modeld.evaluate(X_test, y_test)
```

```python
y_pred = modeld.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test, y_pred))
```
